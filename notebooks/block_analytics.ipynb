```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sys
import os

module_path = os.path.abspath('../') # root directory
if module_path not in sys.path:
    sys.path.append(module_path)

DATA_PATH  = os.path.abspath('../data')
```

```python
cols_to_read = [
    "height",
    "gas_used_percentage",
    "burnt_fees_percentage",
    "timestamp",
]

flr_data = pd.read_csv(
    os.path.join(DATA_PATH, "flare_blocks_20250522.csv"),
    usecols=cols_to_read,
    dtype={
        "height": int,
        "gas_used_percentage": float,
        "burnt_fees_percentage": float,
    },
    parse_dates=["timestamp"],
    date_format="%Y-%m-%dT%H:%M:%S.%fZ"
)

sgb_data = pd.read_csv(
    os.path.join(DATA_PATH, "songbird_blocks_20250522.csv"),
    usecols=cols_to_read,
    dtype={
        "height": int,
        "gas_used_percentage": float,
        "burnt_fees_percentage": float,
    },
    parse_dates=["timestamp"],
    date_format="%Y-%m-%dT%H:%M:%S.%fZ"
)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,4))

ax[0].hist(flr_data["gas_used_percentage"], bins=50, density=True)
ax[1].hist(sgb_data["gas_used_percentage"], bins=50, density=True)

ax[0].set_title("Gas Used Distribution (Flare)")
ax[1].set_title("Gas Used Distribution (Songbird)")

ax[0].set_xlabel("Gas Used Percentage (%)")
ax[1].set_xlabel("Gas Used Percentage (%)")
ax[0].set_ylabel("Density")
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
from plotly.offline import plot


# Create Plotly line plot
fig = px.line(
    sgb_data,
    x="timestamp",
    y="gas_used_percentage",
    title="Gas Used Percentage Over Time",
    labels={"timestamp": "Timestamp", "gas_used_percentage": "Gas Used Percentage (%)"}
)

# Add range slider to x-axis
fig.update_layout(xaxis=dict(rangeslider=dict(visible=True)))

# Export to standalone HTML file
output_file = os.path.join(DATA_PATH, "sgb_output.html")
plot(fig, filename=output_file, auto_open=False)

output_file
```
